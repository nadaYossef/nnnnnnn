### Handling Outliers

Outliers can significantly affect model performance. We'll use box plots to visualize the distribution of numerical features and identify potential outliers. After identification, we can decide on a strategy, such as capping, transformation, or removal, based on the nature of the data and the extent of the outliers.

```python
# Identify numerical columns for outlier detection
numerical_cols_for_outliers = [col for col in numerical_cols if col not in ['transaction_id', 'user_id', 'addicted_label']]

# Create box plots for numerical features to visualize outliers
plt.figure(figsize=(18, 15))
for i, col in enumerate(numerical_cols_for_outliers):
    plt.subplot(3, 3, i + 1) # Adjust subplot grid as needed
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.ylabel('') # Remove y-label as it's redundant
plt.tight_layout()
plt.show()
```

From the box plots, we can observe that some features like `daily_screen_time_hours`, `social_media_hours`, `gaming_hours`, `work_study_hours`, `notifications_per_day`, `app_opens_per_day`, and `weekend_screen_time` show some data points outside the 'whiskers' which are generally considered outliers. `age` and `sleep_hours` appear to be well-behaved.

Given the nature of screen time and notification data, these 'outliers' might represent legitimate high-usage users rather than data entry errors. Removing them could lead to loss of valuable information about extreme usage patterns. Capping them (e.g., to the 99th percentile) might be a more suitable approach to reduce their influence without losing data entirely. For simplicity and given the goal of understanding addiction, we will proceed without explicitly capping them at this stage, as these extreme values could be indicative of addiction. However, if these values caused significant model issues, capping or transformation would be the next step.

## 3. Feature Engineering

We will create new features that combine existing information or represent new insights into user behavior. These include ratios and aggregated measures that can better highlight patterns related to smartphone usage and potential addiction.

```python
# Calculate Average Daily Screen Time
# Assuming 'daily_screen_time_hours' is for weekdays and 'weekend_screen_time' is for 2 weekend days
df_cleaned['average_daily_screen_time'] = (df_cleaned['daily_screen_time_hours'] * 5 + df_cleaned['weekend_screen_time'] * 2) / 7

# Handle potential division by zero for ratio features
df_cleaned['daily_screen_time_hours_adjusted'] = df_cleaned['daily_screen_time_hours'].replace(0, np.nan)
df_cleaned['sleep_hours_adjusted'] = df_cleaned['sleep_hours'].replace(0, np.nan)

# Productivity Ratio
df_cleaned['productivity_ratio'] = df_cleaned['work_study_hours'] / df_cleaned['daily_screen_time_hours_adjusted']

# Social Media Usage Percentage (of total daily screen time)
df_cleaned['social_media_ratio'] = df_cleaned['social_media_hours'] / df_cleaned['daily_screen_time_hours_adjusted']

# Gaming Usage Percentage (of total daily screen time)
df_cleaned['gaming_ratio'] = df_cleaned['gaming_hours'] / df_cleaned['daily_screen_time_hours_adjusted']

# Notifications per hour of screen time
df_cleaned['notifications_per_screen_hour'] = df_cleaned['notifications_per_day'] / df_cleaned['daily_screen_time_hours_adjusted']

# App opens per hour of screen time
df_cleaned['app_opens_per_screen_hour'] = df_cleaned['app_opens_per_day'] / df_cleaned['daily_screen_time_hours_adjusted']

# Screen Time vs. Sleep Ratio
df_cleaned['screen_time_sleep_ratio'] = df_cleaned['daily_screen_time_hours_adjusted'] / df_cleaned['sleep_hours_adjusted']

# Fill any NaN values created by division by zero with 0 (or a suitable value)
df_cleaned[['productivity_ratio', 'social_media_ratio', 'gaming_ratio',
            'notifications_per_screen_hour', 'app_opens_per_screen_hour', 'screen_time_sleep_ratio']] = \
    df_cleaned[['productivity_ratio', 'social_media_ratio', 'gaming_ratio',
                'notifications_per_screen_hour', 'app_opens_per_screen_hour', 'screen_time_sleep_ratio']].fillna(0)

# Drop the adjusted columns
df_cleaned = df_cleaned.drop(columns=['daily_screen_time_hours_adjusted', 'sleep_hours_adjusted'])

print("New features created. Displaying head of DataFrame with new features:")
display(df_cleaned.head())
```

## 8. Data Leakage Check

Data leakage occurs when information from outside the training dataset is used to create the model. This can lead to a model that performs exceptionally well on the training and test data but poorly on new, unseen data. A common source of leakage is features that are direct proxies for the target variable or that contain information that would not be available at prediction time.

In our case, the `addiction_level` feature is highly suspicious. It was used to impute missing values and was identified as one of the most important features. It is very likely a direct or indirect measure of whether a person is addicted, making it a strong candidate for data leakage into our `addicted_label` (binary 0/1) target variable. Let's examine its relationship with the target closely.

```python
# Examine the relationship between 'addiction_level' and 'addicted_label'
# Group by 'addiction_level' and check the mean of 'addicted_label'
# (which represents the proportion of 'addicted' individuals for each level)

leakage_check = df_cleaned.groupby('addiction_level')['addicted_label'].value_counts(normalize=True).unstack().fillna(0)

print("Proportion of 'addicted_label' (0/1) for each 'addiction_level':")
display(leakage_check)

# Visualize this relationship
plt.figure(figsize=(10, 6))
sns.heatmap(leakage_check, annot=True, cmap='viridis', fmt='.2f', linewidths=.5)
plt.title('Relationship between Addiction Level and Addicted Label')
plt.xlabel('Addicted Label')
plt.ylabel('Addiction Level')
plt.show()

print("\nThis clearly shows a very strong, almost deterministic, relationship between 'addiction_level' and 'addicted_label'. This is a strong indication of data leakage. To prevent this, 'addiction_level' should be removed from the features for training a predictive model for 'addicted_label'.")
```

### Addressing Data Leakage

As identified, `addiction_level` presents a significant data leakage issue. To build a robust and generalizable model, we must remove `addiction_level` from our feature set before training. We will modify the feature selection step to explicitly exclude this column.

```python
from scipy.stats import pearsonr

# Create an empty list to store results
correlation_p_values = []

# Iterate through each numerical column (excluding the target itself)
for col in numerical_cols:
    if col != 'addicted_label':
        # Drop rows with NaN values for the specific column and target for accurate correlation calculation
        # although `df_cleaned` should not have NaNs in these columns after imputation and feature engineering.
        temp_df = df_cleaned[[col, 'addicted_label']].dropna()

        # Calculate Pearson correlation and p-value
        correlation, p_value = pearsonr(temp_df[col], temp_df['addicted_label'])
        correlation_p_values.append({'Feature': col, 'Correlation': correlation, 'P-value': p_value})

# Convert the list of dictionaries to a DataFrame for better display
correlation_p_values_df = pd.DataFrame(correlation_p_values)

# Sort by absolute correlation to see strongest relationships first
correlation_p_values_df['Abs_Correlation'] = correlation_p_values_df['Correlation'].abs()
correlation_p_values_df = correlation_p_values_df.sort_values(by='Abs_Correlation', ascending=False).drop(columns='Abs_Correlation')

print("Statistical Significance of Correlations with 'addicted_label':")
display(correlation_p_values_df)
```
